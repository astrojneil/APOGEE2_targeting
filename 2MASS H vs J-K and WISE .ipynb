```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from astropy.io import ascii
from astropy.table import Table
%matplotlib inline

#read in the necessary data
catalogWise = ascii.read('../../WISEandGALEXmatchCatalogs/WISE/wiseMatchesCatalog.txt')
catalog2MASS = ascii.read('../../WISEandGALEXmatchCatalogs/2MASS/2MASSmatches.tbl')

controlL = ascii.read('../../WISEandGALEXmatchCatalogs/LowExtWiseand2MASS.tbl.txt')
controlH = ascii.read('../../WISEandGALEXmatchCatalogs/HighExtWiseand2MASS.tbl.txt')
```

```python
#read in the magnitudes needed
w1 = np.array(catalogWise['W1mag'])
w2 = np.array(catalogWise['W2mag'])
w3 = np.array(catalogWise['W3mag'])
w4 = np.array(catalogWise['W4mag'])

massJ = np.array(catalog2MASS['j_m'])
massH = np.array(catalog2MASS['h_m'])
massK = np.array(catalog2MASS['k_m'])


w1ctrlL = np.array(controlL['w1mpro'])
w2ctrlL = np.array(controlL['w2mpro'])
w3ctrlL = np.array(controlL['w3mpro'])
w4ctrlL = np.array(controlL['w4mpro'])

w1ctrlH = np.array(controlH['w1mpro'])
w2ctrlH = np.array(controlH['w2mpro'])
w3ctrlH = np.array(controlH['w3mpro'])
w4ctrlH = np.array(controlH['w4mpro'])

massJctrlL = np.array(controlL['j_m_2mass'])
massHctrlL = np.array(controlL['h_m_2mass'])
massKctrlL = np.array(controlL['k_m_2mass'])

massJctrlH = np.array(controlH['j_m_2mass'])
massHctrlH = np.array(controlH['h_m_2mass'])
massKctrlH = np.array(controlH['k_m_2mass'])
```

```python
#2MASS plot
plt.scatter(massJctrlL-massKctrlL, massJctrlL, marker='.', color = 'blue')
plt.scatter(massJctrlH-massKctrlH, massHctrlH, marker='.', color = 'green')
plt.scatter(massJ-massK,massH, marker='x', color = 'magenta')
plt.xlabel('J-K', fontsize = 20)
plt.ylabel('H mag', fontsize = 20)
plt.xlim(-1, 8)
plt.ylim(19,1) #fainter objects near bottom
plt.title('2MASS H vs J-K', fontsize = 20)
plt.legend(('Low Extinction', 'High Extinction', 'YSO'), loc=1, fontsize = 20)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
fig.savefig('../../WISEandGALEXmatchCatalogs/Images/2MASSHvJ-K.png', dpi=100)
```

```python
#WISE plot
plt.scatter(w1ctrlL-w3ctrlL, w2ctrlL, marker='.', color = 'blue')
plt.scatter(w1ctrlH-w3ctrlH, w2ctrlH, marker='.', color = 'green')
plt.scatter(w1-w3, w2, marker='x', color = 'magenta')
plt.xlabel('W1-W3', fontsize = 20)
plt.ylabel('W1', fontsize = 20)
plt.xlim(-1,10)
plt.ylim(18,-1) #fainter objects near bottom
plt.title('WISE W2 vs W1-W3', fontsize = 20)
plt.legend(('Low Extinction', 'High Extinction', 'YSO'), loc=1, fontsize = 20)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
fig.savefig('../../WISEandGALEXmatchCatalogs/Images/WISEW2vW1-W3.png', dpi=100)
```

```python
#2MASS color vs color
plt.scatter(massHctrlL-massKctrlL, massJctrlL - massHctrlL, marker='.', color = 'blue')
plt.scatter(massHctrlH-massKctrlH, massJctrlH - massHctrlH, marker='.', color = 'green')
plt.scatter(massH-massK, massJ - massH, marker='x', color = 'magenta')
plt.xlabel('H-K', fontsize = 20)
plt.ylabel('J-H', fontsize = 20)
plt.xlim(-2, 5)
plt.ylim(-2,6)
plt.title('2MASS J-H  vs H-K', fontsize = 20)
plt.legend(('Low Extinction', 'High Extinction', 'YSO'), loc=1, fontsize = 20)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
fig.savefig('../../WISEandGALEXmatchCatalogs/Images/2MASSJ-HvH-K.png', dpi=100)
```

```python
#WISE color vs color
plt.scatter(w3ctrlL-w4ctrlL, w1ctrlL - w2ctrlL, marker='.', color = 'blue')
plt.scatter(w3ctrlH-w4ctrlH, w1ctrlH - w2ctrlH, marker='.', color = 'green')
plt.scatter(w3-w4, w1 - w2, marker='x', color = 'magenta')
plt.xlabel('W3-W4', fontsize = 20)
plt.ylabel('W1-W2', fontsize = 20)
plt.xlim(-1,7)
plt.ylim(-2,4)
plt.title('WISE W1 - W2 vs W3-W4', fontsize = 20)
plt.legend(('Low Extinction', 'High Extinction', 'YSO'), loc=2, fontsize = 20)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
fig.savefig('../../WISEandGALEXmatchCatalogs/Images/WISEW1-W2vW3-W4.png', dpi=100)
```
